$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5 + x + 4/x

def df(x):
    return 1 - 4/x**2


critical_points = [x for x in np.roots([4, -4, 1]) if x > 0]


def d2f(x):
    return 8/x**3

x_vals = np.linspace(0.1, 5, 400)  
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = 5 + x + \\frac{4}{x}$')
plt.scatter(critical_points, f(np.array(critical_points)), color='red', label='Puntos Críticos')

for x in critical_points:
    second_derivative = d2f(x)
    if second_derivative > 0:
        plt.text(x + 0.1, f(x) + 0.5, f'Mínimo', color='blue')
    else:
        plt.text(x + 0.1, f(x) - 0.5, f'Máximo', color='green')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de $f(x)$ con Puntos Críticos y Máximos/Mínimos')
plt.legend()
plt.grid(True)
plt.show()

print("Puntos críticos:", critical_points)
```
